```python
import pandas as pd
from tqdm import tqdm
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
import os
import random
import numpy as np

SEED = 3000

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True

seed_everything(SEED)
```

## Model

```python
access_token = ''

model_name = "KillerShoaib/gemma-2-9b-bangla-4bit"
tokenizer = AutoTokenizer.from_pretrained(model_name, token=access_token)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map="auto",
    token=access_token,
)
```

```python
Eng = []

for i in tqdm(range(len(df))):
    problem1 = '$\\frac{1+3+5+\\dots+(2n+1)}{\\frac{1}{2} + \\frac{1}{6} + \\frac{1}{12} + \\dots + \\frac{1}{342}} = 342$. $n$ এর মান নির্ণয় করো।'
    ans1 = 'Find the value of n such that $\\frac{1+3+5+\\dots+(2n+1)}{\\frac{1}{2} + \\frac{1}{6} + \\frac{1}{12} + \\dots + \\frac{1}{342}} = 342$.'

    problem2 = 'a এর কোন মানের জন্য \\(x^2 + 4x + 8y + 4xy + 4y^2 + axyz\\) বিভাজ্য হবে \\(x + 2y\\) দ্বারা?'
    ans2 = 'For what value of a will the expression \\(x^2 + 4x + 8y + 4xy + 4y^2 + axyz\\) be divisible by \\(x + 2y\\)?'

    problem3 = '$1 - \\left(\\frac{3!}{2! \\times 1! \\times 2^1}\\right) + \\left(\\frac{4!}{2! \\times 2! \\times 2^2}\\right) - \\left(\\frac{5!}{2! \\times 3! \\times 2^3}\\right) + \\dots$ ধারাটির অসীমতক সমষ্টিকে $\\frac{m}{n}$ আকারে প্রকাশ করা গেলে, $m \\times n$ এর মান নির্ণয় করো। (যেখানে $m$, $n$ পরস্পর সহমৌলিক)'
    ans3 = '$1 - \\left(\\frac{3!}{2! \\times 1! \\times 2^1}\\right) + \\left(\\frac{4!}{2! \\times 2! \\times 2^2}\\right) - \\left(\\frac{5!}{2! \\times 3! \\times 2^3}\\right) + \\dots$ If the infinite sum of the series is expressed in the form $\\frac{m}{n}$, find the value of $m \\times n$. (where $m$, $n$ are relatively prime)'

    problem4 = '$2^{20} \times 3^{10} \times 5^{8}$ সংখ্যাটির কতগুলো পূর্ণ ঘন উৎপাদক আছে?'
    ans4 = 'How many perfect cube factors does the number $2^{20} \times 3^{10} \times 5^{8}$ have?'

    problem5 = 'অন্তপুর একটি বিশাল এলাকা। সেখানে পানির ট্যাঙ্কের মাধ্যমে পানি সরবরাহ করা হয়। এলাকার মাঝে একটি বড় ট্যাঙ্ক আছে, বাকিগুলো ছোট। বড় ট্যাঙ্ক থেকে শুরু করে ছোট ট্যাঙ্ক, সেগুলো থেকে অন্য ট্যাঙ্ক - এভাবে ক্রমান্বয়ে পানি সরবরাহ হয়। সিস্টেমটি এমনভাবে সাজানো হয়েছে যেন একটি ট্যাঙ্কে (বড় ট্যাঙ্ক বাদে) শুধু অন্য একটি ট্যাঙ্ক থেকেই পানি আসে; আর একটি ট্যাঙ্ক হয় অন্য 7 টি ট্যাঙ্কে পানি সরবরাহ করে, নাহয় কাউকেই করেনা। এলাকায় মোট 2024 টি ট্যাঙ্ক থাকলে, এমন কয়টি ট্যাঙ্ক আছে যেগুলো পানি সরবরাহ করে না?'
    ans5 = 'Antopur is a vast area. Water is supplied through water tanks there. There is a large tank in the middle of the area, the rest are small. Water is supplied gradually from the large tank to the small tanks, from those to other tanks. The system is arranged so that a tank (excluding the large tank) receives water from only one other tank; and one tank supplies water to 7 other tanks, no one else does. If there are a total of 2024 tanks in the area, how many tanks do not supply water?'

    problem6 = '\\(f(n) = 3f(n + 1) - 2f(n - 1)\\), \\(f(0)=0\\) এবং \\(f(1)=1\\) হলে, 2f(2023) + 3f(2024) এর মান নির্ণয় কর।'
    ans6 = 'If \\(f(n) = 3f(n + 1) - 2f(n - 1)\\), \\(f(0)=0\\) and \\(f(1)=1\\). Find the value of 2f(2023) + 3f(2024).'

    problem7 = 'দুইটি ধনাত্মক পূর্ণসংখ্যার গুণফল 15, সংখ্যা দুইটির গ.সা.গু 1 হলে, সংখ্যা দুইটির ল.সা.গু নির্ণয় কর।'
    ans7 = 'Determine the least common multiple of two positive integers whose product is 15 and whose greatest common divisor is 1.'

    problem8 = '$\\lceil {\\frac{n+(n+1)}{n+2}} \\rceil + \\lceil {\\frac{(n + 1) + (n + 2)}{(n + 3)}} \\rceil + \\lceil {\\frac{(n + 2) + (n + 3)}{n + 4}} \\rceil + \\dots + \\lceil {\\frac{(n+m-2)+(n+m-1)}{n+m}} \\rceil =301$, যেখানে $0 < n < 1$। $m$ এর মান নির্ণয় কর। \n$\\lceil \\rceil$ দ্বারা সিলিং ফাংশন বুঝায়, যার মান কোনো দশমিক সংখ্যার পরবর্তী পূর্ণসংখ্যা বুঝায়। যেমন, $\\left\\lceil {2.1} \\right\\rceil = 3, \\left\\lceil {3} \\right\\rceil = 3, \\left\\lceil {2.6} \\right\\rceil = 3$'
    ans8 = '$\\lceil {\\frac{n+(n+1)}{n+2}} \\rceil + \\lceil {\\frac{(n + 1) + (n + 2)}{(n + 3)}} \\rceil + \\lceil {\\frac{(n + 2) + (n + 3)}{n + 4}} \\rceil + \\dots + \\lceil {\\frac{(n+m-2)+(n+m-1)}{n+m}} \\rceil =301$, where $0 < n < 1$. Determine the value of $m$. \n$\\lceil \\rceil$ symbol denotes the ceiling function, which gives the next integer after a decimal number. For example, $\\left\\lceil {2.1} \\right\\rceil = 3, \\left\\lceil {3} \\right\\rceil = 3, \\left\\lceil {2.6} \\right\\rceil = 3$'

    problem = df.iloc[i]['Problem']
    messages = [
            {
                "role": "user", 
                "content": f"Please translate the given sentences to complete English sentences and put your translated lines altogether within \\boxed{{}} in the last line. Here are the sentences containing Bengali and English words:\n{problem1}\nPlease do translation only."
            },
            {
                "role": "system", 
                "content": ans1
            },
            {
                "role": "user", 
                "content": f"Please translate the given sentences to complete English sentences and put your translated lines altogether within \\boxed{{}} in the last line. Here are the sentences containing Bengali and English words:\n{problem2}\nPlease do translation only."
            },
            {
                "role": "system", 
                "content": ans2
            },
            {
                "role": "user", 
                "content": f"Please translate the given sentences to complete English sentences and put your translated lines altogether within \\boxed{{}} in the last line. Here are the sentences containing Bengali and English words:\n{problem3}\nPlease do translation only."
            },
            {
                "role": "system", 
                "content": ans3
            },
            {
                "role": "user", 
                "content": f"Please translate the given sentences to complete English sentences and put your translated lines altogether within \\boxed{{}} in the last line. Here are the sentences containing Bengali and English words:\n{problem4}\nPlease do translation only."
            },
            {
                "role": "system", 
                "content": ans4
            },
            {
                "role": "user", 
                "content": f"Please translate the given sentences to complete English sentences and put your translated lines altogether within \\boxed{{}} in the last line. Here are the sentences containing Bengali and English words:\n{problem5}\nPlease do translation only."
            },
            {
                "role": "system", 
                "content": ans5
            },
            {
                "role": "user", 
                "content": f"Please translate the given sentences to complete English sentences and put your translated lines altogether within \\boxed{{}} in the last line. Here are the sentences containing Bengali and English words:\n{problem6}\nPlease do translation only."
            },
            {
                "role": "system", 
                "content": ans6
            },
            {
                "role": "user", 
                "content": f"Please translate the given sentences to complete English sentences and put your translated lines altogether within \\boxed{{}} in the last line. Here are the sentences containing Bengali and English words:\n{problem7}\nPlease do translation only."
            },
            {
                "role": "system", 
                "content": ans7
            },
            {
                "role": "user", 
                "content": f"Please translate the given sentences to complete English sentences and put your translated lines altogether within \\boxed{{}} in the last line. Here are the sentences containing Bengali and English words:\n{problem8}\nPlease do translation only."
            },
            {
                "role": "system", 
                "content": ans8
            },
            {
                "role": "user", 
                "content": f"Please translate the given sentences to complete English sentences and put your translated lines altogether within \\boxed{{}} in the last line. Here are the sentences containing Bengali and English words:\n{problem}\nPlease do translation only."
            },
        ]
    input_tensor = tokenizer.apply_chat_template(messages, add_generation_prompt=True, return_tensors="pt")
    outputs = model.generate(input_tensor.to(model.device),
                             max_new_tokens=1024, 
                             temperature=0.7,
                             top_p=0.7,
                             do_sample=True)
    result = tokenizer.decode(outputs[0][input_tensor[0].shape[0]:], skip_special_tokens=True)
    Eng.append(result)
```
